```python
driver = webdriver.Chrome('C:/Users/Rohan/Desktop/ISB/Data Collection/chromedriver_win32/chromedriver.exe') 
```

identified #search as css selector and sending the search key to webdriver youtube site for searching

```python
Modi_TOI_pages = "https://timesofindia.indiatimes.com/blogs/?s=modi"
driver.get(Modi_TOI_pages) 
n=1
Itr=500
while (n < Itr):
    print(n)
    try:
        n += 1
        print(Modi_TOI_pages)
        reviews = driver.find_elements_by_xpath('.//h2[@class="media-heading"]//a')
        review_list = [item.text for item in reviews]
        review_dt   = driver.find_elements_by_css_selector('.date')
        review_date = [item.text for item in review_dt]
        review_Year = [re.split(', ', item2,maxsplit=0)[-2] for item2 in review_date]
        review_df = pd.DataFrame(data=[review_list,review_Year],index=['Reviews','Year']).transpose()
        data_train= data_train.append(review_df,ignore_index=True)
        Modi_TOI_pages = "https://timesofindia.indiatimes.com/blogs/page/" + str(n) + "/?s=modi"
        driver.get(Modi_TOI_pages)
    except:
        n = Itr             

driver.close()
```

```python
data_train_recent = data_train[data_train['Year'].isin(['2018','2017','2016','2015'])]
data_train_recent.to_csv("indiatoday_blogs_26.csv",index=False,encoding='utf_8')
```
