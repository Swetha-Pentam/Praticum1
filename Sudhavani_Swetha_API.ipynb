3. Follwoing is the code  Fetch the recent atrticles published about "business analytics",
from all the sources. As each call using API returning 100 articles, page_size is given 100 otherwise it is 20 default.
The API is called iteratively changing the page number from 1 to... until exception is received from newsAPI. The exception given by newsAPI after capturing 999 articles and for more it requires paid subscription. Hence the loop is stopped and catpured 999 articles

initial one time call to get total count of articles (max count) and loop accordingly

```python
#google-news-in - 03
#the-times-of-india - 02
#the-hindu 01
def getrecent(j):
    result={}
    try:
        result = newsapi.get_everything(q='modi',
                                        sort_by='publishedAt',
                                        page=j,
                                        page_size=100,
                                        sources='the-times-of-india'
                                      )         
    except:
        print ("Exception received from newsAPI as End")
        
    return result
```

Further loop established as below until exception is received and appended the results to dataframe as below. The results of First call above are first added to dataframe and then it is append.
API call gives the results in JSON format hence they are converte from Dict form to dataframe 

```python
df_articles = pd.DataFrame()
i=100
j=1
No_of_Articles = 101

while (i < No_of_Articles):
    print (i,j)
    fresult = getrecent(j)
    if fresult:
        all_articles = fresult['articles']
        temp = pd.DataFrame.from_dict(all_articles)
        df_articles = df_articles.append(temp,ignore_index=True)
        if j==1:
            No_of_Articles = fresult['totalResults']
        i = i+100
        j = j+1
    else:
        i = No_of_Articles              
 
```
